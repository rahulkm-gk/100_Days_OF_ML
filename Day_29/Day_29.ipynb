```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
X_train, X_test , y_train, y_test = train_test_split(df.drop(columns=["Survived"]),df["Survived"],test_size=0.2,random_state=42)
```

```python
# Simple imputation apply for fill the null value


si_age = SimpleImputer()
si_embraked = SimpleImputer(strategy="most_frequent")

X_train_age = si_age.fit_transform(X_train[["Age"]])
X_train_embrked = si_embraked.fit_transform(X_train[["Embarked"]])

X_test_age = si_age.transform(X_test[["Age"]])
X_test_embraked = si_embraked.transform(X_test[["Embarked"]])
```

```python
X_train_embrked
```

```python
X_train.head(2)
```

```python
# One hot encoding
ohe_sex = OneHotEncoder(
    sparse_output=False,
    handle_unknown="ignore"
)
ohe_embarked = OneHotEncoder(
    sparse_output=False,
    handle_unknown="ignore"
)

X_train_sex = ohe_sex.fit_transform(X_train[["Sex"]])
X_train_embrked = ohe_embarked.fit_transform(X_train_embrked)

X_test_sex = ohe_sex.transform(X_test[["Sex"]])
X_test_embraked = ohe_embarked.transform(X_test_embraked)
```

```python
X_train_sex
```

```python
X_train_embrked
```

```python
X_train_rem = X_train.drop(columns=["Sex","Age","Embarked"])
```

```python
X_test_rem = X_test.drop(columns=["Sex","Age","Embarked"])
```

```python
X_train_transform = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embrked),axis=1)
X_test_transform = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embraked),axis=1)
```

```python
X_test_transform.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transform,y_train)
```

```python
y_pred = clf.predict(X_test_transform)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
